### Plotting data on the amazon rainforest between the years 2004 -2019 in different amazon areas

```python
import pandas as pd
```

```python
data = pd.read_csv('amazon_data.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig_size = plt.rcParams["figure.figsize"]
print("Current size:", fig_size)
fig_size[0] = 12
fig_size[1] = 20
plt.rcParams["figure.figsize"] = fig_size


plt.title('Deforestation area (km) by year in mutiple Amazonian states')
plt.xlabel('Year')
plt.ylabel('Deforestation area (km)')

plt.plot( 'Ano/Estados', 'AC', data=data, marker='', color='skyblue', linewidth=2, label="Acre")
plt.plot( 'Ano/Estados', 'AM', data=data, marker='', color='olive', linewidth=2, label="Amazonas")
plt.plot( 'Ano/Estados', 'AP', data=data, marker='', color='red', linewidth=2, label="Amapa")
plt.plot( 'Ano/Estados', 'MA', data=data, marker='', color='green', linewidth=2,label="Maranhao")
plt.plot( 'Ano/Estados', 'MT', data=data, marker='', color='magenta', linewidth=2, label="Matto Groso")
plt.plot( 'Ano/Estados', 'PA', data=data, marker='', color='pink', linewidth=2, label="Para")
plt.plot( 'Ano/Estados', 'RO', data=data, marker='', color='orange', linewidth=2, label="Rondonia")
plt.plot( 'Ano/Estados', 'RR', data=data, marker='', color='yellow', linewidth=2, label="Roraima")
plt.plot( 'Ano/Estados', 'TO', data=data, marker='', color='indigo', linewidth=2, label="Tocantins")
plt.plot( 'Ano/Estados', 'AMZ LEGAL', data=data, marker='', color='black', linewidth=2, label="Full Amazon")
plt.legend(title = 'State',fancybox=True,loc=1, fontsize='large')
```
